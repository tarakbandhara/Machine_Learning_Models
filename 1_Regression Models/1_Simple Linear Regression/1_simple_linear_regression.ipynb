==>>Simple Linear Regression :

Importing Libraries :

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Importing the Dataset:

```python
dataset = pd.read_csv("Salary_Data.csv")
x = dataset.iloc [:, :-1].values

y = dataset.iloc [:, -1].values
```

Splitting dataset into training set and test set:

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

Training the simple regression model on the training set:

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

Predicting test set results

```python
y_pred = regressor.predict(x_test) 
```

Visulaizing the training sets results : #x is years of experience and y is the salary

```python
# x is years of experience and y is the salary

plt.scatter(x_train, y_train, color = "red")
plt.plot(x_train, regressor.predict(x_train), color = "blue")
plt.title("Salary vs Experience(Training Set)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

Visulaizing the test sets results : #x is years of experience and y is the salary

```python
# x is years of experience and y is the salary

plt.scatter(x_test, y_test, color = "red")
plt.plot(x_train, regressor.predict(x_train), color = "blue")  #this is regression line which will be same for both test and training set as it is derived from formula so it here won't matter if we change x_train with x_test or not.
plt.title("Salary vs Experience(Training Set)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```
