=====> Random Forest Regression : --->

===> Importing Libraries : --->

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

===> Importing the Dataset : --->

```python
dataset = pd.read_csv("Position_Salaries.csv")
x = dataset.iloc[:, 1:2].values
y = dataset.iloc[:,2].values
```

===>Training the Random Forest Regression model on whole Dataset : --->

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators= 10, random_state= 0)  # estimators in n_estimators are the number of trees used to train/make random forest regression model.
regressor.fit(x , y)
```

===>Predicting new result : --->

```python
regressor.predict([[6.5]])
```

===>Visualizing the random forest regression results (High resolution) : --->

```python
x_grid = np.arange(min(x), max(x), 0.01)
x_grid = x_grid.reshape((len(x_grid), 1))
plt.scatter(x , y , color = "red")
plt.plot(x_grid, regressor.predict(x_grid), color = "blue")
plt.title("Truth or Bluff (Random Forest Regression)")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```
